### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# let's define fit weather conditions for fishing ;) : cloudiness == 30, Max Temp ==30 to 16, Wind Speed == 3, Humidity == 10 to 45
df.describe()
```

```python
# •Drop any rows will null values.
df_fit = df.loc[(df["Cloudiness"] <= 30)&(df["Max Temp"] <= 30)&(df["Max Temp"] >= 16)&(df["Humidity"] <= 45)&(df["Humidity"] >= 10)&(df["Wind Speed"] <= 10)&(df["Humidity"] >= 5)]

df_fit
```

```python
# •Store into variable named hotel_df.
hotel = df_fit.loc[:, ["City", "Country", "Lat", "Lon"]]

#•Add a "Hotel Name" column to the DataFrame.
hotel["Hotel Name"] = ""
hotel
```

```python
type()
```

```python
#Set parameters to search for hotels with 5000 meters. Preparing for iterrows method

params = {
    "type": "hotel",
    "keyword": "hotel",
    "radius": 5000,
    "key": g_key    
}

#•Hit the Google Places API for each city's coordinates.
url_base = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
```

```python
# Store the first Hotel result into the DataFrame.
for index, row in hotel.iterrows():
    lat=row["Lat"]
    lon=row["Lon"]
    city=row["City"]
    
    params["location"] = f"{lat},{lon}"
    
    print(f"Results for index {index}:{city}")
    
    response = requests.get(url_base, params=params).json()
# functions for stracting    
    results = response['results']
    
    try:
        print(f"The hotel suggested in {city} is {results[0]['name']}.")
        hotel.loc[index, "Hotel Name"] = results[0]['name']

    # if there is no hotel available, show missing field
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------ Next----------------")
    
# Print end of search once searching is completed
print("""
____________________________________________________
                
                WE FINISH!
____________________________________________________
""")
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel.iterrows()]
locations = hotel[["Lat", "Lon"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content = hotel_info)
markers
# Display Map
fig.add_layer(markers)

fig
```
